```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.poisson(lam=0.998, size=455), hist=False, label='455 case')
sns.distplot(random.poisson(lam=0.998, size=1455), hist=False, label='1455 case')
sns.distplot(random.poisson(lam=0.998, size=225), hist=False, label='255 case')
plt.axvline(x=2)
plt.grid()
plt.show()
```

As the number of words increased the variace also increased drastically.
We can conclude that the number of words is proportional to variance.
